```python
# PCA with Multiple Classifiers

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Wine.csv')
X = dataset.iloc[:, 0:13].values  # Features (columns 0 to 12)
y = dataset.iloc[:, 13].values    # Target variable (column 13)

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Applying PCA (Principal Component Analysis)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  # Reducing to 2 components for visualization
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio for each component:", explained_variance)

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
log_classifier = LogisticRegression(random_state=0)
log_classifier.fit(X_train, y_train)

# Predicting the Test set results
log_y_pred = log_classifier.predict(X_test)

# Fitting SVM (Linear Kernel) to the Training set
from sklearn.svm import SVC
svc_classifier = SVC(kernel='linear', random_state=0)
svc_classifier.fit(X_train, y_train)

# Predicting the Test set results for SVM
svc_y_pred = svc_classifier.predict(X_test)

# Fitting KNN to the Training set
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)

# Predicting the Test set results for KNN
knn_y_pred = knn_classifier.predict(X_test)

# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

# Predicting the Test set results for Naive Bayes
nb_y_pred = nb_classifier.predict(X_test)

# Fitting Decision Tree to the Training set
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_classifier.fit(X_train, y_train)

# Predicting the Test set results for Decision Tree
dt_y_pred = dt_classifier.predict(X_test)

# Fitting Random Forest to the Training set
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
rf_classifier.fit(X_train, y_train)

# Predicting the Test set results for Random Forest
rf_y_pred = rf_classifier.predict(X_test)

# Making the Confusion Matrix and Accuracy for each model
from sklearn.metrics import confusion_matrix, accuracy_score

# Create a helper function to display confusion matrix and accuracy
def print_metrics(y_test, y_pred, model_name):
    cm = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Confusion Matrix for {model_name}:\n{cm}")
    print(f"Accuracy for {model_name}: {accuracy:.4f}\n")
    return cm, accuracy

# Print metrics for each classifier
print_metrics(y_test, log_y_pred, "Logistic Regression")
print_metrics(y_test, svc_y_pred, "SVM (Linear Kernel)")
print_metrics(y_test, knn_y_pred, "KNN")
print_metrics(y_test, nb_y_pred, "Naive Bayes")
print_metrics(y_test, dt_y_pred, "Decision Tree")
print_metrics(y_test, rf_y_pred, "Random Forest")

# Function for visualization (to avoid duplication)
def visualize_results(X_set, y_set, classifier, title):
    from matplotlib.colors import ListedColormap
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                         np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
    plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue')))
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
    plt.title(title)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.legend()
    plt.show()

# Visualising the results for each model (Training set)
visualize_results(X_train, y_train, log_classifier, 'Logistic Regression (Training set)')
visualize_results(X_train, y_train, svc_classifier, 'SVM (Linear Kernel) (Training set)')
visualize_results(X_train, y_train, knn_classifier, 'KNN (Training set)')
visualize_results(X_train, y_train, nb_classifier, 'Naive Bayes (Training set)')
visualize_results(X_train, y_train, dt_classifier, 'Decision Tree (Training set)')
visualize_results(X_train, y_train, rf_classifier, 'Random Forest (Training set)')

# Visualising the results for each model (Test set)
visualize_results(X_test, y_test, log_classifier, 'Logistic Regression (Test set)')
visualize_results(X_test, y_test, svc_classifier, 'SVM (Linear Kernel) (Test set)')
visualize_results(X_test, y_test, knn_classifier, 'KNN (Test set)')
visualize_results(X_test, y_test, nb_classifier, 'Naive Bayes (Test set)')
visualize_results(X_test, y_test, dt_classifier, 'Decision Tree (Test set)')
visualize_results(X_test, y_test, rf_classifier, 'Random Forest (Test set)')
```
